```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20)
```

```python
class yourmodel:
    
    def __init__(self):
        self.m = None
        self.b = None
        
    def fit(self,xtrain,ytrain):
        num =0
        den = 0
        
        for i in range(xtrain.shape[0]):
            #now calculate m value formula = (xi - xmean) (yi - ymean)
                                                    #-------------------------
                                                     # (xi - xmean)2
                    #first (xi - xmean) - (yi - y mean)
                    
                num = num + ((xtrain[i] - xtrain.mean()) * (ytrain[i] - ytrain.mean()))
                 #first (xi - xmean) - (yi - y mean)
                    
           
                    # send (xi - xmean)2
                    
                den =  den + ((xtrain[i] - xtrain.mean()) * (xtrain[i] - xtrain.mean()))
            
        self.m = num/den
        #now calculate  b
        # b =  y(mean)- m xmean
        
        self.b = ytrain.mean() - (self.m * xtrain.mean())
        
        print("coef",self.m)
        print("intercept",self.b)
            
    def predict(self,xtest):
          # y = mx + b 
        print(xtest)
        return self.m * xtest +self.b
            
        # complete your model now fit and predict tha model
```

```python
lr  = yourmodel()
```

```python
lr
```

```python
#now fit tha model
lr.fit(xtrain,ytrain)
```
